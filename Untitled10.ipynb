```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("WcWtsi.csv")
print(data)
```

```python
x=data[['YearsExperience']]
y=data['Salary']
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
predicted_Score=model.predict([[3]])
print("Predicted score for 3 hours of study: ",predicted_Score[0])
```

```python
plt.scatter(x,y,color='blue',label='Actual data')
plt.plot(x,model.predict(x),color='green',label='Regresssion Line')
plt.xlabel('YearsOfExperience')
plt.ylabel("Salary")
plt.title("StudyExam vs score")
plt.legend()
plt.show()
```

```python
slope=model.coef_
intercept=model.intercept_
print(slope)
print(intercept)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
```
